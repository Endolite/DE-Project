```python
#import required libraries for model creation
import numpy as np
import matplotlib.pyplot as plt
```

```python
def geometric_brownian_motion(num_steps, delta_t, initial_value, drift, volatility):
    # Generate a geometric Brownian motion path
    time = np.arange(0, (num_steps * delta_t), delta_t)
    increments = np.random.normal(0, np.sqrt(delta_t), num_steps)
    path = initial_value * np.exp(np.cumsum((drift - 0.5 * volatility**2) * delta_t + volatility * np.sqrt(delta_t) * increments))
    
    return time, path
```

```python
# Model parameters
num_steps = 1000
delta_t = 0.01
initial_value = 100  # Initial value of the process
drift = 0.3         # Drift coefficient
volatility = 0.7     # Volatility coefficient
```

```python
# Generate and plot the geometric Brownian motion
time, path = geometric_brownian_motion(num_steps, delta_t, initial_value, drift, volatility)
plt.plot(time, path, label='Google')
plt.title('Stock Price Over Time')
plt.xlabel('Time (Days past 11/20/2022)')
plt.ylabel('Price ($)')
plt.legend()
plt.show()
```
